```python
import os
import pandas as pd

# Google Drive 내의 폴더 경로 설정
folder_path = '/content/drive/MyDrive/project/data/data2/'  # 폴더 경로를 수정하세요

# 폴더 내 모든 CSV 파일 합치기
all_files = [os.path.join(folder_path, f) for f in os.listdir(folder_path) if f.endswith('.csv')]

# 각 파일 데이터를 읽어서 하나의 데이터프레임으로 합치기
combined_df = pd.concat([pd.read_csv(file) for file in all_files], ignore_index=True)

# 결과 확인
print("합쳐진 데이터프레임:")
print(combined_df.head())

# 합쳐진 데이터 저장
output_path = '/content/drive/MyDrive/project/data/combined_data.csv'
combined_df.to_csv(output_path, index=False)
print(f"합쳐진 데이터가 저장되었습니다: {output_path}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 데이터 로드
travel_df = pd.read_csv(
    "/content/drive/MyDrive/combined_data.csv",
    encoding="utf-8-sig",
    sep=',',
    index_col=None
)

# 컬럼 이름 설정
travel_df.columns = ['SEQ_NO', 'ALL_KWRD_RANK_CO', 'SRCHWRD_NM', 'UPPER_CTGRY_NM', 'LWPRT_CTGRY_NM',
                     'MOBILE_SCCNT_VALUE', 'PC_SCCNT_VALUE', 'SCCNT_SM_VALUE', 'SCCNT_DE']

# 특정 값 (20000 이상) 데이터 필터링
filtered_df = travel_df[travel_df["SCCNT_SM_VALUE"] >= 10000]

# 그래프 생성
plt.figure(figsize=(20, 9))
bars = plt.bar(filtered_df["SRCHWRD_NM"], filtered_df["SCCNT_SM_VALUE"], color='green')

# 그래프 설정
plt.title("축제검색수 10000이상")
plt.xlabel("SRCHWRD_NM")
plt.ylabel("SCCNT_SM_VALUE")
plt.xticks(rotation=45)  # x축 레이블 회전
plt.tight_layout()

plt.show()
```
