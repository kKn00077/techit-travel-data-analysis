```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: 데이터 로드
file_path = '/content/drive/MyDrive/project/data/20241108163329_성·연령별방문자_연령별.csv'
df = pd.read_csv(file_path, encoding='euc-kr')

# Step 2: 성별 방문자 수 합계 및 비율 계산
gender_counts = df.groupby('남성 방문자 비율')['여성 방문자 비율'].sum()
gender_percentage = gender_counts / gender_counts.sum() * 100

# Step 3: 파이 차트 생성
plt.figure(figsize=(8, 8))
wedges, texts, autotexts = plt.pie(
    gender_percentage,
    labels=gender_counts.index,
    autopct='%.1f%%',
    startangle=90,
    colors=['lightblue', 'pink'],
    textprops={'fontsize': 14},
    wedgeprops={'edgecolor': 'black'}
)

# 제목 추가
plt.title("성별 방문자 비율", fontsize=16)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 제공된 데이터를 기반으로 DataFrame 생성
data = {
    "연령": ["70세 이상", "60~69세", "50~59세", "40~49세", "30~39세", "20~29세", "10~19세", "0~9세"],
    "남성 방문자 비율": [4.0, 7.7, 9.0, 8.7, 7.9, 6.1, 5.8, 0.6],
    "여성 방문자 비율": [4.4, 8.1, 9.5, 8.6, 7.3, 6.3, 5.5, 0.7],
}

# DataFrame 생성
df = pd.DataFrame(data)

# 남성과 여성 비율을 합산하여 전체 비율 계산
df["전체 방문자 비율"] = df["남성 방문자 비율"] + df["여성 방문자 비율"]

# 파이 차트 생성
plt.figure(figsize=(10, 6))
plt.pie(
    df["전체 방문자 비율"],
    labels=df["연령"],
    autopct="%1.1f%%",
    startangle=90,
    colors=plt.cm.Paired.colors,
    wedgeprops={"edgecolor": "black"},
)
plt.title("연령대별 전체 방문자 비율", fontsize=16)
plt.show()
```
