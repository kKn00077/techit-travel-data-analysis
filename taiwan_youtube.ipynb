```python
from googleapiclient.discovery import build
from googleapiclient.errors import HttpError
from oauth2client.tools import argparser

API_KEY = 'AIzaSyBOwcj2N-LBi7gcc30TKuXtz4Kb8VyHI3o'
YOUTUBE_API_SERVICE_NAME = 'youtube'
YOUTUBE_API_SERVICE_VERSION = 'v3'

youtube = build(YOUTUBE_API_SERVICE_NAME, YOUTUBE_API_SERVICE_VERSION, developerKey = API_KEY)
```

```python
query = "臺灣旅行韓國慶典"

results = []

response = youtube.search().list(
    q=query,
    part="snippet",
    type="video",
    maxResults=50
).execute()

results.extend(response["items"])

# 페이지 반복 (최대 5,000개까지)
while "nextPageToken" in response and len(results) < 5000:
    response = youtube.search().list(
        q=query,
        part="snippet",
        type="video",
        maxResults=50,
        pageToken=response["nextPageToken"]
    ).execute()
    results.extend(response["items"])



print(f"Total results fetched: {len(results)}")
for item in results:
    print(item["snippet"]["title"], item["id"]["videoId"])
```
